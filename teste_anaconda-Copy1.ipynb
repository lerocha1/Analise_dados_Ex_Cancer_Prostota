```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
Dados = pd.read_csv('Prostate_cancer.csv',sep = ',')
Dados.head(10)
```

```python
# Vamos traduzir as colunas para melhor entedimento. Na pratica saiu do ingles para o Portugues!
Dados.rename( columns={
    'diagnosis_result' : 'Diagnóstico resultado',
    'radius' : 'Raio',
    'texture' : 'Textura',
    'perimeter' : 'Perímetro',
    'area' : 'Área',
    'smoothness' : 'Suavidade' ,
    'compactness' : 'Compacidade',
    'symmetry' : 'Simetria',
    'fractal_dimension' : 'Dimensão do fragmento'
}, inplace=True )

Dados.head(60)
```

```python
#praticamente um summary do R em python, de acordo com os dados!
Dados.describe()
```

```python
#TRocando siglas do diagnostico pelas palavras propriamente ditas!
Dados['Diagnóstico resultado'] = Dados['Diagnóstico resultado'].replace('M', 'Maligno')
Dados['Diagnóstico resultado'] = Dados['Diagnóstico resultado'].replace('B', 'Benigno')
Dados.head(10)
```

```python
#Removendo coluna do id!
Dados.drop(['id'], axis=1, inplace=True)
```

```python
Dados.head(10)
```

```python
#Plot grafico para percentagem de Maligno e Benigno
# Calculando os valores
Valores = Dados['Diagnóstico resultado'].value_counts()
Valores
```

```python
Labels = [Valores.index.tolist()]
Labels
```

```python
plt.figure (figsize=(8,6) ) #tamanho do grafico, só ele nao plota nada
plt.pie(Valores,
        labels = Labels[0],
        shadow=True,
        explode=(0.0, 0.1),
        startangle=90,
        autopct='%1.1f%%' );#se esquecer os dois %%
```

```python
from sklearn import preprocessing


Ajuste_Variavel = preprocessing.LabelEncoder()

Dados['Diagnóstico resultado'] = Ajuste_Variavel.fit_transform( Dados['Diagnóstico resultado'] )
```

```python
Correlacao = Dados.corr()

plt.figure(figsize=(14,10)) #tamanho do grafico, só ele nao plota nada

sns.heatmap(Correlacao, linewidths=.1, cmap='viridis', annot=True)
```

```python
#Plotando os histogramas para cada Coluna da base de dados após fit e Correlação!
sns.set( font_scale=1.3, rc={'figure.figsize':(15, 15)}) #parametros do gráfico


#Plot dos histogramas
ax = Dados.hist(bins=20, color='blue')

```

```python
#Dividindo/separando a base de dados!
Caracterisitica = Dados.iloc[:,1:].values
Previsor = Dados.iloc[:,0:1].values

#funcao para separar os dados de treino e teste
from sklearn.model_selection import train_test_split

x_treino, x_teste, y_treino, y_teste = train_test_split(Caracterisitica, 
                                                        Previsor,
                                                        test_size=0.20,
                                                        random_state=10)
print('Dados de Treinamento:', len(x_treino))
print('Dados de Teste:', len(x_teste))
```

```python
#Treinando o modelo
#importando função de Naive-Bayes - Calcula a probabilidade a "posteori" para eventuais acontecimentos!

from sklearn.naive_bayes import GaussianNB

nb = GaussianNB() #armazena a função no objeto "nb"

nb.fit(x_treino, y_treino)

print('P Score do Modelo:',nb.score(x_treino, y_treino))
```

```python
#usando Previsões para os resultados obitidos

Previsoes = nb.predict(x_teste)
Previsoes
```

```python
#Agora faremos previsões atraves da matriz de confusão em função do y_teste para com o Previsões do predict de x_teste, 
#com Naive-Bayes

from sklearn.metrics import confusion_matrix, classification_report

mcf = confusion_matrix(y_teste, Previsoes) #armazena mcf (matriz confusão), conforme escrito acima!

#configura tamanho do gráfico
plt.figure(figsize=(10,6))

#mapa de calor
sns.heatmap(mcf,linewidths=.1, annot=True)
```

```python
plt.figure(figsize=(10,6)) #sempre definir tamanhao antes de plotar grafico, para não sair zoado, fora de ecala!
#plotando mesmo grafico com viridis para quem tem Dautonismo!!
sns.heatmap(mcf, linewidths=.05, cmap='viridis', annot=True)
```

```python
#Analise do Report com função classification_report, importada anteriormente in 37. 

print(classification_report(y_teste, Previsoes))

```
